```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


plt.figure(figsize=(10, 6))

# Histogram with finer bins and adjusted range
ax = sns.histplot(
    df['generated_power_kw'],
    bins=50,  # Increased bin count for finer detail
    kde=True,  # Include KDE for a smoother distribution
    kde_kws={'bw_adjust': 2},  # Adjust KDE bandwidth for better fit
    stat="density", # Display as probability density to ensure area under curve is 1
    color='skyblue', # Change color for better visualization
)

# Add a vertical line for the mean
plt.axvline(
    df['generated_power_kw'].mean(),
    color='red',
    linestyle='dashed',
    linewidth=2,
    label=f"Mean: {df['generated_power_kw'].mean():.2f}"
)

# Add a vertical line for the median
plt.axvline(
    df['generated_power_kw'].median(),
    color='green',
    linestyle='dashed',
    linewidth=2,
    label=f"Median: {df['generated_power_kw'].median():.2f}"
)

# Add text annotations for quartiles
q1 = np.quantile(df['generated_power_kw'], 0.25)
q3 = np.quantile(df['generated_power_kw'], 0.75)

ax.text(q1, 0.005, f"Q1: {q1:.2f}", ha='center', va='bottom', color='black')
ax.text(q3, 0.005, f"Q3: {q3:.2f}", ha='center', va='bottom', color='black')


plt.title('Distribution of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Density') # Update y-axis label to reflect probability density
plt.legend() # Show legend for mean and median lines
plt.show()
```
